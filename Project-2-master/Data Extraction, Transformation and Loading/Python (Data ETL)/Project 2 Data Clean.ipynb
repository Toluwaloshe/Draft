```python
#Import all dependencies
import pandas as pd
from sqlalchemy import create_engine
```

```python
#Locate all files
GDP_file = 'Raw Data/GDP Annual Growth.xls'
workforce_file = 'Raw Data/Female Percentage of Total Labor Force.xls'
lat_long_file = 'Raw Data/Country Lat Long.xlsx'
```

## Clean GDP Data

```python
#Read GDP data
GDP_data = pd.read_excel(GDP_file, sheet_name='Data')
GDP_data.head()
```

```python
#Remove the first 2 rows as they do not contain data
GDP_data_2 = GDP_data.drop(GDP_data.index[[0,1]])
GDP_data_2
```

```python
#Make firt row the header
new_header = GDP_data_2.iloc[0] #grab the first row for the header
GDP_data_3 = GDP_data_2[1:] #take the data less the header row
GDP_data_3.columns = new_header #set the header row as the df header

GDP_data_3
```

```python
#Dropping "Indicator Name" and "Indicator Code" because we don't need it and "1960" and "2019" because there's no data there
GDP_data_4 = GDP_data_3.drop(columns = ['Indicator Name', 'Indicator Code', 1960, 2019])
GDP_data_4
```

```python
#Renaming the Country Code and Country Name columns
GDP_data_renamed = GDP_data_4.rename(columns={'Country Name': 'country_name', 'Country Code':'country_code'})

GDP_data_renamed
```

```python
#Making Country Code the index so that rows are deleted based on the code
indexed = GDP_data_renamed.set_index('country_code')

indexed
```

```python
#Creating a function to drop rows that are not countries
def drop(indexed):
    dropped = indexed.drop(['ARB', 'CEB', 'CSS', 'EAP', 'EAR', 'EAS', 'ECA', 'ECS', 'EMU', 'EUU', 'FCS', 'FSM', 'HIC', 'HKG', 'HPC', 'IBD', 'IBT', 'IDA', 'IDB', 'IDX', 'INX', 'KGZ', 'KNA', 'KOR', 'LAC', 'LCN', 'LDC', 'LIC', 'LMC', 'LMY', 'LTE', 'MAC', 'MAF', 'MEA', 'MIC', 'MKD', 'MNA', 'NAC', 'OED', 'OSS', 'PRE', 'PRK', 'PSS', 'PST', 'SAS', 'SSA', 'SSD', 'SSF', 'SST', 'STP', 'SVK', 'SWZ', 'SXM', 'TEA', 'TEC', 'TLA', 'TMN', 'TSA', 'TSS', 'UMC', 'VCT', 'VIR', 'WLD'])
    
    return dropped
```

```python
#Check if rows were dropped
GDP_data_dropped = drop(indexed)

GDP_data_dropped
```

```python
#Making country code its own column again
GDP_data_dropped['code'] = GDP_data_dropped.index

GDP_data_dropped
```

```python
#Create a function to unpivot/create a year column and a column with GDP growth and replace NaN values with 0

def convert(GDP_data_dropped):
    converted = pd.melt(GDP_data_dropped, id_vars=['country_name', 'code'], 
                        var_name='year', value_name='gdp_growth')
    #Sort values by country name and year
    converted_df = converted.sort_values(by=["country_name", "year"])
    #round GDP growth
    converted_df_2 = converted_df.round({"gdp_growth":3, "year":0})
    converted_df_3 = converted_df_2.fillna(0)
    
    return converted_df_3
```

```python
GDP_data_5 = convert(GDP_data_dropped)

GDP_data_5
```

## Clean Female Workforce Participation Data

```python
#Read workforce data
workforce_data = pd.read_excel(workforce_file, sheet_name='Data')
workforce_data.head()
```

```python
#Remove the first 2 rows as they do not contain data
workforce_data_2 = workforce_data.drop(workforce_data.index[[0,1]])
workforce_data_2
```

```python
#Make firt row the header
new_header_2 = workforce_data_2.iloc[0] #grab the first row for the header
workforce_data_3 = workforce_data_2[1:] #take the data less the header row
workforce_data_3.columns = new_header_2 #set the header row as the df header

workforce_data_3.head()
```

```python
#Dropping "Indicator Name" and "Indicator Code" because we don't need it, "1960" because there's no data there, and "2019" because GDP data does not have 2019 data
workforce_data_4 = workforce_data_3.drop(columns = ['Indicator Name', 'Indicator Code', 1960, 2019])
workforce_data_4
```

```python
#Renaming the Country Code and Country Name columns
workforce_data_renamed = workforce_data_4.rename(columns={'Country Name': 'country_name', 'Country Code':'country_code'})

workforce_data_renamed
```

```python
#Making Country Code the index so that rows are deleted based on the code
indexed_2 = workforce_data_renamed.set_index('country_code')

indexed_2
```

```python
#Check if rows were dropped
workforce_data_dropped = drop(indexed_2)

workforce_data_dropped
```

```python
#Making country code its own column again
workforce_data_dropped['code'] = workforce_data_dropped.index

workforce_data_dropped
```

```python
workforce_data_5 = convert(workforce_data_dropped)

workforce_data_5
```

```python
#Renaming gdp_growth to % of total laborforce
workforce_data_6 = workforce_data_5.rename(columns={'gdp_growth': 'fem_laborforce'})

workforce_data_6
```

## Clean Country Lat Long Data

```python
lat_long_data = pd.read_excel(lat_long_file)

lat_long_data.head()
```

```python
latlong_data_1 = lat_long_data.rename(columns={'latitude': 'lat', 'longitude':'long', 'country': 'code', 'name': 'country'})

latlong_data_1
```

```python
#Re-ordering the columns
latlong_data_2 = latlong_data_1[['country', 'code', 'lat', 'long']]

latlong_data_2
```
